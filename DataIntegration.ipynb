```python
# Import libraries
import pandas as pd

# Load the dataset
df = pd.read_csv("gender_submission.csv.csv")

# Display first 5 rows
print("First 5 Rows:")
print(df.head())

# Dataset information
print("\nDataset Info:")
print(df.info())

# Check shape
print("\nShape of Dataset:")
print(df.shape)

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Check duplicate rows
print("\nDuplicate Rows:", df.duplicated().sum())

# Remove duplicates (if any)
df = df.drop_duplicates()

# Check data types
print("\nData Types:")
print(df.dtypes)

# Convert data types (optional)
df["PassengerId"] = df["PassengerId"].astype(int)
df["Survived"] = df["Survived"].astype(int)

# Check unique values
print("\nUnique Values:")
print(df.nunique())

# Check statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Detect outliers using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))

print("\nNumber of Outliers:")
print(outliers.sum())

# Save cleaned dataset
df.to_csv("gender_submission_cleaned.csv", index=False)

print("\nData preprocessing completed successfully!")
```

```python
import matplotlib.pyplot as plt

plt.hist(df["PassengerId"], bins=10)
plt.title("Histogram of PassengerId")
plt.xlabel("PassengerId")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Survived"].value_counts().plot(kind="bar")

plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
df["Survived"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Survival Percentage")
plt.ylabel("")
plt.show()
```

```python
plt.plot(df["PassengerId"], df["Survived"])

plt.title("PassengerId vs Survived")
plt.xlabel("PassengerId")
plt.ylabel("Survived")
plt.show()
```
